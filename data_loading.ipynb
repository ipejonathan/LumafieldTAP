## Generate Point Cloud Representations from STL Files

```python
import open3d as o3d
```

```python
mesh = o3d.io.read_triangle_mesh(r"C:\Users\Jonathan Ipe\Dropbox\Coding\LumafieldTAP\stl_data\geode\geode-dense.stl")

mesh.compute_vertex_normals()
pcd = mesh.sample_points_uniformly(number_of_points=100000)

o3d.io.write_point_cloud("./point_cloud_data/geode/geode-dense.ply", pcd)

# For visualization
o3d.visualization.draw_geometries_with_editing([pcd])
```
